```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# 예시 logit 값들
logits = [-0.316, -0.353, -0.400]

probs = [sigmoid(l) for l in logits]

print("Logits:", logits)
for i in range(len(probs)):
    print("Probabilities:", probs[i] * 100)
```
